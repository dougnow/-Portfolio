### Manipulacion de Columnas de Fechas

```python
df1.info()
# De las columnas fechas, extraiga el mes y el a√±o segun lo visto
```

```python
# el a√±o ser√≠a de la siguiente forma
df1['Year']=pd.to_datetime(df1['Order Date']).dt.year
```

```python
# el mes ser√≠a de la siguiente manera
df1['mes']=pd.to_datetime(df1['Order Date']).dt.month_name()
```

**Realice una exploracion del Dataset** 

**Resumen de informacion, agrupaciones, tablas pivote**

```python
# Realizce una exploracion de los datos y brinde algun criterio de la informacion

# primero , conocer sobre los clientes y como est√°n comprando
clientes = df1.loc[:,['CustomerName','Amount']]
clientes1 = clientes.sort_values('Amount',ascending=False).groupby('CustomerName').head(3).reset_index(drop=True)
clientes1
```

```python
clientes2=df1.pivot_table(index=['CustomerName'],values=['Amount','Profit'],
                          aggfunc='sum').sort_values(by=['Amount','Profit'],ascending=False)
clientes2
```

```python
# El cliente con mayor suma de Amount corresponde 1er=Yaanvi  2ndo= Pooja 3ero= Abhishek
# esto segun la informacion extraida de las agrupaciones 
# y tabla dinamicas (clientes, clientes1, clientes2)
```

```python
# segundo segun las categorias y las ventas relacionadas
categorias=df1.pivot_table(index=['Category'],values=['Amount','Profit'],
                          aggfunc='sum').sort_values(by=['Amount','Profit'],ascending=False)
categorias
```

```python
# La categoria que genera mayor venta es Electronics, la mayor profit es clothing
# segun la tabla pivote categorias

# C√≥mo est√°n las ventas seg√∫n los meses?? 
ventas_mes=df1.pivot_table(index=['mes'],values=['Amount','Profit'],
                          aggfunc='sum').sort_values(by=['Amount','Profit'],ascending=False)
ventas_mes
```

```python
# los meses de mayores ventas corresponde a Enero, Marzo y noviembre
# segun la tabla pivote Ventas_mes

# Las ventas por region 
region_ventas=df1.pivot_table(index=['City'],values=['Amount','Profit'],
                          aggfunc='sum').sort_values(by=['Amount','Profit'],ascending=False)
region_ventas
```
