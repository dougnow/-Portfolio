## 6- Busqueda Binaria


```python
# comparar la eficiencia de naive search   y luego con una busqueda binaria
import random 
import time # para realizar conteo de tiempo

def busqueda_ingenua(lista, objetivo):# creAR UN ALISTA Y EN ESA LISTA SABER SI EL VALOR SE ENCUENTRA OBJETIVO/TARGET
    for i in range(len(lista)): #range(len(lista)) va a recorrer de 0,1,2,3   por ejemplo recorrer el indice
        if lista[i] == objetivo:
            return i   # si se cumple retornar el valor de i
    return -1  # sino se cumple regresa valor -1
```

### Implementar el algoritmo de busqueda binaria para mejorar el tiempo de busqueda

1- la lista debe estar ordenada de forma ascendente de lo contrario no funciona

2- agregar un limite inferior y un limite superior

```python
def busqueda_binaria(lista, objetivo, limite_inferior=None, limite_superior=None):
    if limite_inferior is None:
        limite_inferior = 0 # inicio de la lista
    if limite_superior is None:
        limite_superior = len(lista)-1 # Final de las lista 
        
    if limite_superior < limite_inferior:
        return -1

    punto_medio = (limite_inferior + limite_superior) //2   # permite buscar el punto medio de la lista
    
    if lista[punto_medio] == objetivo:
        return punto_medio
    elif objetivo < lista[punto_medio]:
        return busqueda_binaria(lista,objetivo, limite_inferior,punto_medio-1)
    else:
        return busqueda_binaria(lista,objetivo, punto_medio+1, limite_superior)
    
    
if __name__=='__main__':
    mi_lista=[1,3,5,10,12]
    print(busqueda_binaria(mi_lista,12))
```
