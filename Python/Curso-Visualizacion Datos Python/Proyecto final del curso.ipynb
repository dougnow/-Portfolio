```python
# primero se procede a seleccionar las columnas de interes
avocado_price1 = avocado_price.iloc[:,[1,2,3,7,8,9,10,11]]
avocado_price1.head()
```

```python
avocado_price1.dtypes  # según se puede ver la columna de fecha está en formato objeto
```

```python
# cambiar la columna Date  de objeto a formato fecha para los uso necesarios 
avocado_price1['Date']= pd.to_datetime(avocado_price1['Date'],  
                                 format='%Y-%m-%d %H:%M:%S')
```

```python
avocado_price1.dtypes # se utiliza este paso para obsevar los tipos de datos de las columnas
```

```python
# una vez relizado el cambio de formato, se puede extraer la inormacion de Dia, Año , Mes-Nombr, y # Mes

avocado_price1['Dia']=pd.to_datetime(avocado_price1['Date']).dt.day_name()
avocado_price1['Anio']=pd.to_datetime(avocado_price1['Date']).dt.year
avocado_price1['Mes_Nomb']=pd.to_datetime(avocado_price1['Date']).dt.month_name()
avocado_price1['Mes']=pd.to_datetime(avocado_price1['Date']).dt.month
```

### Etapa 5 y 6 - Simplificar  y Visualiar Datos -Dataset Avacado

```python
# precio promedio por tipo de aguacate
price1 = pd.DataFrame(avocado_price1.groupby(['Date','type'])[['AveragePrice']].mean()).reset_index()
price1
```

```python
# Se muestra histograma segun el comportamiento de los precios segun el tipo de agucate
sns.histplot(price1['AveragePrice'], kde= True, color = 'gold') # con la opcion kde= true permite ver la linea
plt.show()
```

```python
# Crear una agrupacion por meses
avocado_mes = pd.DataFrame(avocado_price1.groupby(['Mes'])[['Total Volume','Total Bags']].sum()).reset_index()
avocado_mes
```

```python
# Se muestra una grafica del comportamiento del volumen de ventas segun los meses
fig, ax = plt.subplots()
ax.plot(avocado_mes['Mes'],avocado_mes['Total Volume'])
plt.show;
```

```python
# Crear una agrupacion por años
avocado_anio = pd.DataFrame(avocado_price1.groupby(['Anio'])[['Total Volume','Total Bags']].sum())
avocado_anio
```

```python
# Se crea una agrupacion por el año 
avocado_anio2 = pd.DataFrame(avocado_price1.groupby(['Date','Anio'])[['Total Volume','Total Bags']].sum()).reset_index()

per_18= avocado_anio2[avocado_anio2['Anio']==2018]
per_17= avocado_anio2[avocado_anio2['Anio']==2017]
avocado_anio2.head()
```

```python
# grafio de dispersion del total de volumen por Total bags

fig, ax = plt.subplots()

ax.scatter(per_17['Total Volume'],per_17['Total Bags'],
          color= 'red', label='2017')
ax.scatter(per_18['Total Volume'],per_18['Total Bags'],
          color= 'blue', label='2018')
plt.show()
```

```python
# Se grefica el comportamiento de las ventas segun volumnen por los meses
fig, ax = plt.subplots()
ax.plot(avocado_mes['Mes'],avocado_mes['Total Volume'])
plt.show;
```

```python
# agrupacion por tipo 
avocado_tipo = pd.DataFrame(avocado_price1.groupby(['type'])[['Total Volume','Total Bags']].sum())
avocado_tipo
```

```python
fig, ax = plt.subplots()

ax.barh(avocado_tipo.index, avocado_tipo['Total Volume'])
plt.show()
```

```python
round(avocado_price1.pivot_table(['Total Volume'],['Anio'],['type'],aggfunc= 'mean',fill_value=0),2)
```

```python
round(avocado_price1.pivot_table(['Total Volume'],['Anio'],['type'],aggfunc= 'mean',fill_value=0),2).plot(kind= 'bar')
plt.title("Volumen de Ventas de Aguacates por tipo")
plt.ylabel("Volumne Promedio")
plt.xlabel("Años")
plt.show()
```

### Etapa 5 y 6 - Simplificar  y Visualiar Datos 

```python
# crear una agrupacion por mes 
per_month = pd.DataFrame(supermarket1.groupby(['Mes'])[['Amount','Quantity']].sum()).reset_index()
per_month
```

```python
# se crea la grafica de la anterior agrupacion 
fig, ax = plt.subplots()
ax.plot(per_month['Mes'],per_month['Amount'])
plt.show()
```

```python
# se crea la sumatpria de todo el año, en este dataset corresponde a un solo año 2015
per_year = pd.DataFrame(supermarket1.groupby(['Anio'])[['Amount','Quantity']].sum()).reset_index()
per_year
```

```python
# se crea una agrupacion por el genero de los clientes
round(supermarket1.pivot_table(['Amount'],['Mes'],['Gender'],aggfunc= 'mean',fill_value=0),2)
```

```python
# se grafica la anterior agrupacion de clientes por genero
# para visualizar comportamiento de compra segun el género
round(supermarket1.pivot_table(['Amount'],['Mes'],['Gender'],aggfunc= 'mean',fill_value=0),2).plot(kind = 'bar' ) 
sns.set(rc={'figure.figsize':(10,10)})
plt.title("Ventas por Genero del Cliente")
plt.ylabel("Ventas Promedio")
plt.xlabel("Meses")
plt.show()
```

```python
# se crea una agrupacion por las lineas de producto existentes
per_Product = pd.DataFrame(supermarket1.groupby(['Product line'])[['Amount','Quantity']].sum())
per_Product
```

```python
# Se crea la visalizavion de la agrupacion anterior
fig, ax = plt.subplots()

ax.barh(per_Product.index, per_Product['Quantity'],color=['forestgreen','blue','turquoise','yellow','gray','red'])
plt.title("Ventas por Líneas de Producto")
plt.show()
```

```python
# se crea un agrupacion para saber las compras como están distribuidas según las ciudades
per_city = pd.DataFrame(supermarket1.groupby(['City'])[['Amount','Quantity']].sum())
per_city
```

```python
#se crea la visualiacion de la agrupacion anterior según las ciudades y sus compras
fig, ax = plt.subplots()

ax.barh(per_city.index, per_city['Amount'],color=['forestgreen','blue','turquoise'])
plt.title("Ventas por Ciudad")
plt.show()
```
